```python
gf = Gramformer(models=1, use_gpu=True)
```

```python
gf.correct('I walk to the store and I bought milk')
```

```python
sentences = [
    'I like for walks', 
    'World is flat', 
    'Red a color', 
    'I wish my Computer was run faster.'   
]
```

```python
from nltk.translate.bleu_score import sentence_bleu
```

```python
for sentence in sentences:
    res = gf.correct(sentence)
    score = sentence_bleu(sentence,list(res)[0])
    print(list(res)[0])
    print(score)
```

```python
import gradio as gr
```

```python
def correct(sentence):
    res = gf.correct(sentence)
    score = sentence_bleu(sentence,list(res)[0])
    return list(res)[0],score
```

```python
app_inputs = gr.inputs.Textbox(lines=2, placeholder="Enter sentence here...")
```

```python
interface = gr.Interface(fn=correct, 
                        inputs=app_inputs,
                         outputs='text', 
                        title='Sup, I\'m Gramformer')
```

```python
interface.launch()
```
